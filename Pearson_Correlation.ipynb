```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
datafile = pd.read_csv("sury.csv")
```

```python
datafile.head()
```

```python
pearsoncorr = datafile.corr(method='pearson')
pearsoncorr
```

```python
plt.figure(figsize=(10, 6), dpi = 600) 
sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            cmap='plasma_r',
            annot=True,
            linewidth=0.5)
```

```python
pearsoncorr.to_csv(index=False)
```

```python
pearsoncorr.to_csv('suryout.csv',index=False)
```
